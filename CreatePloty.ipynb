```python
### Import packages
from dash import Dash, html, dash_table, dcc
import pandas as pd
import plotly.express as px

### Incorporate data
labInfo = pd.read_csv(r'/Users/gnickles/Desktop/UWMadison-FungalSupergroup/LabInfo_Sheet_Updated2023.txt', sep='\t')
labInfo.head()

# comine columns at index 3 through 5 into a single columns called department using the melt function
labInfo = pd.melt(labInfo, 
                  id_vars=['Group Leader Name', 'Group Leader Email', 'Institutional Affilitation', "Grad Programs 1", "Grad Programs 2", 'Grad Programs 3', 'Grad Programs 4', 'Grad Programs 5', 'Grad Programs 6', 'Grad Programs 7'], 
                  value_vars=['Department Name 1', 'Department Name 2'], 
                  var_name='Department_Type', 
                  value_name='Department')

labInfo.drop(columns=['Department_Type'], inplace=True)

#now combining the 7 graduate program columns into a single column called graduate program
labInfo = pd.melt(labInfo, 
                  id_vars=['Group Leader Name', 'Group Leader Email', 'Institutional Affilitation', 'Department'], 
                  value_vars=['Grad Programs 1', 'Grad Programs 2', 'Grad Programs 3', 'Grad Programs 4', 'Grad Programs 5', 'Grad Programs 6', 'Grad Programs 7'], 
                  var_name='Graduate_Program_Type', 
                  value_name='Graduate_Program')

labInfo.drop(columns=['Graduate_Program_Type'], inplace=True)

#sorting the values by the group leader name column
labInfo.sort_values(by=['Group Leader Name'], inplace=True)
labInfo.reset_index(drop=True, inplace=True)

### Initializing the app
app = Dash(__name__)

### App layout
app.layout = html.Div([
    html.Div(children='UW-Madison Fungal Supergroup Professors'),
    dash_table.DataTable(data=labInfo.to_dict('records'), page_size=10)
])

### Run the app
if __name__ == '__main__':
    app.run(debug=True)
```
